# K-Means
## Elbow method & Alternate method for labels and cluster centers of clusters.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_blobs
x, y = make_blobs(n_samples = 800, centers = [[8,4], [2,9], [-4,-6], [-1, 1], [5,-9], [-7, 7]], cluster_std = 1)

sns.scatterplot(x[:,0], x[:,1])
```

```python
from sklearn.cluster import KMeans

ss = []
for i in range(1,16):
    kmeans =KMeans(n_clusters = i, init = 'k-means++')
    kmeans.fit(x)
    
    ss.append(kmeans.inertia_)

#px.line(range(1,16), ss)
#plt.plot(range(1,16), ss)
sns.lineplot(range(1,16), ss)
```

```python
kmeans = KMeans(n_clusters= 6, init = 'k-means++', n_init= 12)
y_ = kmeans.fit_predict(x)
```

```python
plt.figure(figsize = (12, 12))
plt.scatter(x[y_==0, 0],x[y_==0, 1], s=100,c='blue')
plt.scatter(x[y_==1, 0],x[y_==1, 1], s=100,c='green')
plt.scatter(x[y_==2, 0],x[y_==2, 1], s=100,c='red')
plt.scatter(x[y_==3, 0],x[y_==3, 1], s=100,c='yellow')
plt.scatter(x[y_==4, 0],x[y_==4, 1], s=100,c='cyan')
plt.scatter(x[y_==5, 0],x[y_==5, 1], s=100,c='magenta')

plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s= 200, c='black')
```

```python
zero = x[y_==0]
one = x[y_==1]
two = x[y_==2]
three = x[y_==3]
four = x[y_==4]
five = x[y_==5]
zero.shape, one.shape, two.shape, three.shape, four.shape, five.shape
```

```python
print(zero[:,0].mean(), zero[:,1].mean())
print(one[:,0].mean(), one[:,1].mean())
print(two[:,0].mean(), two[:,1].mean())
print(three[:,0].mean(), three[:,1].mean())
print(four[:,0].mean(), four[:,1].mean())
print(five[:,0].mean(), five[:,1].mean())
```
